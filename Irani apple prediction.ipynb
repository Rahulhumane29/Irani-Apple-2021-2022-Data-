```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel("C:\\Users\\rahul\\OneDrive\\Desktop\\Apple\Data.xlsx")
```

```python
df
```

```python
X = df["Year"].values.reshape(-1, 1)
y = df["Value"].values
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
year_2023 = np.array([[2023]])
```

```python
prediction_2023 = model.predict(year_2023)
```

```python
print(f"Predicted Export for 2023: {prediction_2023[0]:.2f}")
```

```python
# Predicted Export 5.53 Million USD in 2023
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.scatter(2023, prediction_2023, marker='x')
plt.xlabel("Year")
plt.ylabel("Export Value")
plt.title("Iran Apple Export Prediction")
plt.show()
```
